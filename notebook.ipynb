```python
from preprocessing import DataGen
images_dir = './data/images'
dg = DataGen()
batch_iter = dg.flow(images_dir)
```

```python
pairs, labels = next(batch_iter)
```

```python
import matplotlib.pyplot as plt

limit = 10
for i, (pair, label) in enumerate(zip(pairs, labels), 1):
    fig, ax = plt.subplots(1,2)
    fig.suptitle(f'Sample with label: {label[0]}', fontsize=16)
    ax[0].set_title('Left input')
    ax[1].set_title('Right input')
    ax[0].imshow(keras.preprocessing.image.array_to_img(pair[0]))
    ax[1].imshow(keras.preprocessing.image.array_to_img(pair[1]))
    if i == limit:
        break
```
